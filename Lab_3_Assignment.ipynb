```python
#Task 1
print("Task 1:\n\n")
class Environment:
  def __init__(self, state):
    self.state = state
  def get_percept(self):
    return self.state
class ReflexAgent:
  def __init__(self):
    pass
  def act(self, percept):
    if(percept == 'Heavy Traffic'):
      return 'Extend Green Time'
    else:
      return 'Normal Green'
def run_agent(agent, env):
  percept = env.get_percept()
  action = agent.act(percept)
  print(f"Percept: {percept} -> Action: {action}")

agent = ReflexAgent()
env = Environment('Heavy Traffic')
run_agent(agent, env)


#Task 2
print("\n\nTask 2:\n\n")
import random

class SmartClassroomAgent:
  def __init__(self):
      self.previous_students_present = None
      self.previous_light_status = "OFF"
      self.current_light_status = "OFF"
  def perceive(self):
      return random.choice(["Yes", "No"])
  def act(self, students_present):
      action = "No Action"
      if(students_present == "Yes" and self.current_light_status == "OFF"):
         self.current_light_status = "ON"
         action = "Lights Turned ON"
      elif(students_present == "No" and self.current_light_status == "ON"):
         self.current_light_status = "OFF"
         action = "Lights Turned OFF"
      return action
  def update_model(self, students_present):
      self.previous_students_present = students_present
      self.previous_light_status = self.current_light_status
  def display_state(self, step, students_present, action):
      print("\nStep:", step)
      print("Students Present:", students_present)
      print("Action:", action)
      print("Current Light Status:", self.current_light_status)

agent = SmartClassroomAgent()
for step in range(1,9):
    students_present = agent.perceive()
    action = agent.act(students_present)
    agent.update_model(students_present)
    agent.display_state(step, students_present, action)

#Task 3
print("\n\nTask 3:\n\n")

class StudyPlannerAgent:
  def __init__(self):
    self.subjects = ["AI", "Math", "Physics"]
    self.completed = []
  def act(self):
    for subject in self.subjects:
      if(subject not in self.completed):
        print("Studying", subject)
        self.completed.append(subject)
  def goal_achieved(self):
    if(len(self.completed) == len(self.subjects)):
      print("Goal Achieved: All subjects completed")

agent = StudyPlannerAgent()
agent.act()
agent.goal_achieved()

#Task 4
print("\n\nTask 4:\n\n")

class RestaurantSelectorAgent:
  def __init__(self):
    self.restaurants = {
      "A": {"Distance": 3, "Rating": 7},
      "B": {"Distance": 5, "Rating": 9}
      }
  def calculate_utilities(self):
    utilities = {}
    for r in self.restaurants:
      utilities[r] = (self.restaurants[r]["Rating"] - self.restaurants[r]["Distance"])
      print("Restaurant", r, "Utility =", utilities[r])
    return utilities
  def select_restaurant(self):
    utilities = self.calculate_utilities()
    selected = min([r for r in utilities if utilities[r] == max(utilities.values())])
    print("\nSelected Restaurant:", selected)

agent = RestaurantSelectorAgent()
agent.select_restaurant()

#Task 5
print("\n\nTask 5:\n\n")

import random
class LearningAgent:
  def __init__(self):
    self.q_values = {"Play": 0, "Rest": 0}
    self.rewards = {"Play": 5, "Rest": 1}
  def choose_action(self):
    return max(self.q_values, key=self.q_values.get)
  def update_q(self, action):
    self.q_values[action] += self.rewards[action]

agent = LearningAgent()
for step in range(1,11):
  action = agent.choose_action()
  reward = agent.rewards[action]
  print("Step", step, ": Action", action, "Reward", reward)
  agent.update_q(action)
  print("Q-table Updated\n")

#Task 6
print("\n\nTask 6:\n\n")

class FirefightingRobot:
  def __init__(self):
    self.rooms = {
        "a": "safe", "b": "safe", "c": "fire",
        "d": "safe", "e": "fire", "f": "safe",
        "g": "safe", "h": "safe", "j": "fire"
    }
    self.path = ["a","b","c","d","e","f","g","h","j"]
  def move_and_extinguish(self):
    for room in self.path:
      if (self.rooms[room] == "fire"):
        print("Room", room, "has fire! Extinguishing...")
        self.rooms[room] = "safe"
      else:
        print("Room", room, "is safe. Moving on...")
      self.display_environment()
    print("Final Status of All Rooms:")
    self.display_environment()
  def display_environment(self):
    for r in ["a","b","c"]:
      print(" ".join(self.rooms[x][0].upper() for x in [r]))
    for r in ["d","e","f"]:
      print(" ".join(self.rooms[x][0].upper() for x in [r]))
    for r in ["g","h","j"]:
      print(" ".join(self.rooms[x][0].upper() for x in [r]))
    print("\n")

robot = FirefightingRobot()
robot.move_and_extinguish()

```
