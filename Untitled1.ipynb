```python
def calculate_average(marks):
  sum = 0
  for i in range(len(marks)):
    sum = sum + marks[i]
  return sum/len(marks)

m = [78, 87, 93, 88, 81, 64, 51, 47, 58, 85]
print("Marks: ", m)
avg = calculate_average(m)
print("Average Marks: ", avg)
```
